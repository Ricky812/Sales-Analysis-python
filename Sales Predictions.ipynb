```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing the Dataset
all_sales=pd.read_excel('superstore_sales.xlsx')
```

```python
#Overall Sales Data Trend
all_sales['order_date'].min()
```

```python
all_sales['order_date'].head()
```

```python
#Getting month year from dataset
all_sales['month_year'] = all_sales['order_date'].apply(lambda x: x.strftime("%y-%m")) sales_trend=all_sales.groupby('month_year').sum()['sales'].reset_index()
```

```python
sales_trend=all_sales.groupby('month_year').sum()['sales'].reset_index()
```

```python
#Setting the Figure size
plt.figure(figsize=(20,7))
plt.plot(sales_trend['month_year'], sales_trend['sales'], color='#b80905')
plt.xticks(rotation='vertical',size=15)
plt.show
```

```python
#Top 10 products by sale
prod_sales=pd.DataFrame(all_sales.groupby('product_name').sum()['sales'])
prod_sales=prod_sales.sort_values('sales', ascending=False)
prod_sales[:10]
```

```python
#Most selling products
most_sell_prod=pd.DataFrame(all_sales.groupby('product_name').sum()['quantity'])
most_sell_prod=most_sell_prod.sort_values('quantity', ascending=False)
most_sell_prod[:10]
```
